```python
normal_files = []
normal_labels = []

for file in os.listdir("TB_Chest_Radiography_Database/Normal/"):
    normal_files.append(file)
normal_labels = ['Normal'] * len(normal_files)
normal_labels
```

```python
len(normal_labels)
```

```python
tub_files = []
tub_labels = []

for file in os.listdir("TB_Chest_Radiography_Database/Tuberculosis/"):
    tub_files.append(file)
tub_labels = ['Tuberculosis'] * len(tub_files)
tub_labels
```

```python
all_files = normal_files + tub_files
all_labels = normal_labels + tub_labels
```

```python
all_labels[4000]
```

```python
plt.figure(figsize=(16, 8))
for i in range(16):
    rand = random.randint(0, len(all_files))
    img = plt.imread(os.path.join('TB_Chest_Radiography_Database', all_labels[rand], all_files[rand]))
    plt.subplot(4, 4, i+1)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(all_labels[rand])
plt.show()
```
